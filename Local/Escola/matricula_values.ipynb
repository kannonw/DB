```python
import bd
import random as r
```

```python
mydb = bd.bd_connector()
mycursor = mydb.cursor()
```

```python
mycursor.execute("select count(*) from alunos")
alunos = mycursor.fetchall()
alunos = int(alunos[0][0])
alunos
```

```python
ano = 2010
dia = 1
for m in range(alunos):

    if m % 15 == 0:
        ano += 1
        dia = 1

    diadata = r.choice([dia, dia+1])

    if diadata > dia:
        somador = r.randint(1,3)
        dia += somador

    data = f"{ano}-{11}-{diadata}"

    mycursor.execute(f"""insert into matricula
    values (null, {m+1}, {r.randint(1,7)}, '{data}');""")
    mydb.commit()

    print(f"values (null, {m+1}, {r.randint(1,7)}, '{data}');")
```
